# Graph Representation


```python
class Graph:
    def __init__(self,dictionary=None):
        if dictionary == None:
            dictionary = {}
        self.dictionary = dictionary
    def add(self,v1,v2):
        if v1 in self.dictionary.keys():
            self.dictionary[v1].append(v2)
            print("Edge added to already existing vertex")
            if v2 in self.dictionary.keys():
                self.dictionary[v2].append(v1)
            else:
                self.dictionary[v2] = [v1]
                       
        else:
            self.dictionary[v1] = [v2]
            print("New Node is created and an Edge added to it")
            if v2 in self.dictionary.keys():
                self.dictionary[v2].append(v1)
            else:
                self.dictionary[v2] = [v1]
        
```

```python
d = { "a" : ["b" , "c" , "d"],
       "b" : ["a" , "e" ],
      "c" : ["a" , "d" ],
       "d" : ["a" , "c" , "e"],
       "e" : ["b" , "d"]
    }
```

```python
g = Graph(d)
g.dictionary
```

```python
g.add("b","d")
```

```python
from collections import defaultdict
class Graph1:
    def __init__(self,dictionary=None):
        if dictionary == None:
            dictionary = defaultdict(list)
        self.dictionary = dictionary
    def add(self,v1,v2):
        if v1 in self.dictionary.keys():
            self.dictionary[v1].append(v2)
            print("Edge added to already existing vertex")
            if v2 in self.dictionary.keys():
                self.dictionary[v2].append(v1)
            else:
                self.dictionary[v2] = [v1]
                       
        else:
            self.dictionary[v1] = [v2]
            print("New Node is created and an Edge added to it")
            if v2 in self.dictionary.keys():
                self.dictionary[v2].append(v1)
            else:
                self.dictionary[v2] = [v1]
        
        
```

```python
d = { "a" : ["b" , "c" , "d"],
       "b" : ["a" , "e" ],
      "c" : ["a" , "d" ],
       "d" : ["a" , "c" , "e"],
       "e" : ["b" , "d"]
    }
```

```python
g = Graph1(d)
g.dictionary
```
