# Graph Traversal


# Breadth First Search Algorithm 
> # Similar to Level Order Traversal in Tree

> # Using Queue




### Time Complexity :: $O(V+E)$
### Space Complexity :: $O(V+E)$

```python
def BFS(d,vertex):
    visited = [vertex]
    queue = [vertex]
    while queue:
        pop = queue.pop(0)
        print(pop)
        for Conn_vertex in d.get(pop):
            if Conn_vertex not in visited:
                visited.append(Conn_vertex)
                queue.append(Conn_vertex)
```

```python
d = {   "a" : ["b","c"],
        "b" : ["a", "d", "g"],
        "c" : ["a", "d" , "e"],
        "d" : ["b", "c", "f"],
        "e" : ["f", "c"],
        "f" : ["d", "e","g"],
         "g" : ["b" , "f"]
               }
```

```python
BFS(d,"a")
```

# Depth for Search
> # Using Stack




### Time Complexity :: $O(V+E)$
### Space Complexity :: $O(V+E)$

```python
def DFS(d,vertex):
    visited = [vertex]
    stack = [vertex]
    while stack:
        pop = stack.pop()
        if pop not in visited:
            visited.append(pop)  
      
        for conn_vertex in d.get(pop):
            if conn_vertex not in visited:
                stack.append(conn_vertex)
    return visited
```

```python
d = {   "a" : ["b","c"],
        "b" : ["a", "g",  "d" ],
        "c" : ["a", "d" , "e"],
        "d" : ["b", "c", "f"],
        "e" : [ "c","f"],
        "f" : [ "e","g","d"],
         "g" : ["b" , "f"]
               }
```

```python
DFS(d,"a")
```
