# Linear Search


```python
def Linear_Search(array , value):
    for i in range(len(array)):
        if array[i] == value:
            return f"{value} is found at index : {i}"
    return "Element Doen't exists"
```

```python
L = [5,9,3,1,2,8,4,7,6]
Linear_Search(L,6)
```

```python
Linear_Search(L,19)
```

# Binary Search


### Time Complexity


```python
def binarySearch(array, value):
    start = 0
    end = len(array)-1
    middle = (start+end)//2
    while (array[middle]!=value) and start<=end:
        if value < array[middle]:
            end = middle - 1
        else:
            start = middle + 1 
        middle = (start+end)//2
    if array[middle] == value:
        return middle
    else:
        return -1
```

```python
custArray = [8, 9, 12, 15, 17, 19, 20, 21, 28]
binarySearch(custArray, 20)
```
