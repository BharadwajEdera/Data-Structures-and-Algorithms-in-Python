# Activity Selection Problem is Greedy Algorithm:
> ### At a given Point of time we take best solution , Freeze it and Never come back
> ### There will be No other best activity other than what we have selected as best at that time
> ### Finally All local Solutions Lead to Global Solution
> ### This approach is called Greedy approach and this algorithm is called Greedy Algorithm



> ### Time COmplexity :: $O(Nlog(N))$
> ### Space COmplexity :: $O(1)$

```python
def Activity_selection(activities):
    activities.sort(key= lambda x : x[2])
    print(activities[0][0])
    prev_index = 0
    curr_index = 1
    while curr_index<len(activities):
        if activities[prev_index][2] < activities[curr_index][1]:
            print(activities[curr_index][0])
            prev_index = curr_index
        curr_index += 1
    
```

```python

activities = [["A1", 0, 6],
              ["A2", 3, 4],
              ["A3", 1, 2],
              ["A4", 5, 8],
              ["A5", 5, 7],
              ["A6", 8, 9]]
Activity_selection(activities)
```

# Coin Change Problem



> ### Time COmplexity :: $O(N)$
> ### Space COmplexity :: $O(1)$

```python
def coin_change(n,l):
    l.sort()
    i = len(l)-1
    coins = 0
    while True:
        if n >= l[i]:
            n -= l[i]
            coins += 1
        if n < l[i]:
            i -= 1
        if n == 0:
            return coins
            
```

```python
l = [1,2,5,20,50,100]
coin_change(257, l)
```

# Fractional Knapsack Problem



> ### Time COmplexity :: $O(Nlog(N))$
> ### Space COmplexity :: $O(1)$

```python
class Item:
    def __init__(self,weight,value):
        self.weight = weight
        self.value = value
        self.density = value/weight
```

```python
item1 = Item(20,100)
item2 = Item(30,120)
item3 = Item(10,60)
Item_List = [item1, item2, item3]
```

```python
def FractionalKnapsack(Item_List,Box_Weight):
    Item_List.sort(key = lambda x : x.density , reverse = True)
    value_count = 0
    i = 0
    while True:
        if Box_Weight == 0:
            return value_count
        if Box_Weight > Item_List[i].weight:
            Box_Weight -= Item_List[i].weight
            value_count += Item_List[i].value
            i += 1
        if Box_Weight < Item_List[i].weight:
            value_count += Item_List[i].value*(Box_Weight/Item_List[i].weight)
            Box_Weight -= Box_Weight
    return value_count
            
            
```

```python
FractionalKnapsack(Item_List,50)
```
