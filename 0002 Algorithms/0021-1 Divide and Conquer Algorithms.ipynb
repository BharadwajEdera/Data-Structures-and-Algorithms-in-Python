# Fibonacci Series


```python
def fib(n):
    if n in (0,1):
        return n
    else:
        return fib(n-1) + fib(n-2)
```

```python
fib(6)
```

# Number Factor


```python
def NumberFactor(n):
    if n in (0,1,2):
        return 1
    elif n==3 :
        return 2
    else:
        b1 =  NumberFactor(n-1)
        b3 =  NumberFactor(n-3)
        b4 =  NumberFactor(n-4)
        return b1+b3+b4
```

```python
NumberFactor(5)
```

# House Robber


```python
def HouseRobbery(Houses,Index):
    if Index >= len(Houses):
        return 0
    else:
        First_Select = Houses[Index] + HouseRobbery(Houses,Index+2)
        First_Reject = HouseRobbery(Houses,Index+1)
        return max(First_Select,First_Reject)
```

```python
Houses = [6,7,1,30,8,2,4]
HouseRobbery(Houses, 0)
```

# Covert One String to Other


```python
def ConvertString(s1,s2,index1,index2):
    if index1 == len(s1):
        return len(s2) - index2
    if index2 == len(s2):
        return len(s1) - index1
    if s1[index1] == s2[index2]:
        return ConvertString(s1,s2,index1 + 1 ,index2 + 1)
    else:
        Delete_operation = 1 + ConvertString( s1 , s2 , index1 , index2 + 1 )
        Insert_operation = 1 + ConvertString( s1 , s2 , index1 + 1 , index2 )
        Replace_operation = 1 + ConvertString( s1 , s2 , index1 + 1 , index2 +1 )
        return min(Delete_operation,Insert_operation,Replace_operation)
```

```python
ConvertString("table", "tbrltt", 0, 0)
```

# Zero One Knapsack Problem


```python
class Item:
    def __init__(self,profit,weight):
        self.weight = weight
        self.profit = profit
```

```python
mango = Item(31, 3)
apple = Item(26, 1)
orange = Item(17, 2)
banana = Item(72, 5)
items = [mango, apple, orange, banana]
```

```python
def ZeroKnapsack(items,Box_weight,index):
    if index >= len(items) or Box_weight<=0 or index<0:
        return 0
    elif items[index].weight <= Box_weight:
        option1 = items[index].profit + ZeroKnapsack( items , Box_weight - items[index].weight , index + 1 )
        option2 = 0 + ZeroKnapsack( items , Box_weight  , index + 1 )
        return max(option1,option2)
    else:
        return 0
```

```python
ZeroKnapsack(items,7,0)
```

# Longest Common Subsequence (LCS)


```python
def findLCS(s1, s2, index1, index2):
    if index1 == len(s1) or index2 == len(s2):
        return 0
    if s1[index1] == s2[index2]:
        return 1 + findLCS(s1, s2, index1+1, index2+1)
    else:
        op1 = findLCS(s1,s2, index1, index2+1)
        op2 = findLCS(s1,s2, index1+1, index2)
        return max(op1, op2)
```

```python
print(findLCS("elephant", "eretpat", 0, 0))
```

# Longest Palindromic Subsequence (LPS)


```python
def findLPS(s, startIndex, endIndex):
    if startIndex > endIndex:
        return 0
    elif startIndex == endIndex:
        return 1
    elif s[startIndex] == s[endIndex]:
        return 2 + findLPS(s, startIndex+1, endIndex-1)
    else:
        op1 = findLPS(s, startIndex, endIndex-1)
        op2 = findLPS(s, startIndex+1, endIndex)
        return max(op1, op2)
```

```python
print(findLPS("ELRMENMET", 0, 8))
```

# Minimum Cost to reach to the Last cell 


```python
def findMinCost(twoDArray, row, col):
    if row == -1 or col == -1:
        return float('inf')
    elif row == 0 and col == 0:
        return twoDArray[0][0]
    else:
        op1 = findMinCost(twoDArray, row-1, col)
        op2 = findMinCost(twoDArray, row, col-1)
        return twoDArray[row][col] + min(op1,op2)
```

```python
TwoDList = [[4,7,8,6,4],
           [6,7,3,9,2],
           [3,8,1,2,4],
           [7,1,7,3,7],
           [2,9,8,9,3]
           ]

print(findMinCost(TwoDList, 4,4))
```

# Number of paths to reach to last cell with given cost


```python
def numberOfPaths(twoDArray, row, col, cost):
    if cost < 0:
        return 0
    elif row == 0 and col == 0:
        if twoDArray[0][0] - cost == 0:
            return 1
        else:
            return 0
    elif row == 0:
        return numberOfPaths(twoDArray, 0, col-1, cost - twoDArray[row][col] )
    elif col == 0:
        return numberOfPaths(twoDArray, row-1, 0, cost - twoDArray[row][col] )
    else:
        op1 = numberOfPaths(twoDArray, row -1, col, cost - twoDArray[row][col] )
        op2 = numberOfPaths(twoDArray, row, col-1, cost - twoDArray[row][col] )
        return op1 + op2
```

```python


TwoDList = [[4,7,1,6],
           [5,7,3,9],
           [3,2,1,2],
           [7,1,6,3]
           ]

print(numberOfPaths(TwoDList, 3,3, 25))
```
