# Single Source Shortest Path Problem


# All Pair Shortest Path Problem


> # We are taking Source vertices as one by one and finding the Shortest path

 # We are Running SSSP problem multiple times(Number of Vertices times)



# Floyd Warshall to solve APSP


# Floyd Warshall  doesn't Catch  negative cycle

### Time Complexity :: $O(V^3)$
### Space Complexity :: $O(V^2)$

```python
INF = 9999
# Printing the solution
def printSolution(nV, distance):
    for i in range(nV):
        for j in range(nV):
            if(distance[i][j] == INF):
                print("INF", end=" ")
            else:
                print(distance[i][j], end="  ")
        print(" ")


def floydWarshall(nV, G):
    distance = G
    for k in range(nV):
        for i in range(nV):
            for j in range(nV):
                distance[i][j] = min(distance[i][j], distance[i][k]+distance[k][j])
    
    printSolution(nV, distance)
```

```python

G = [[0, 8, INF,1],
    [INF, 0, 1,INF],
    [4, INF, 0,INF],
    [INF, 2, 9,1]
    ]

floydWarshall(4, G)
```
