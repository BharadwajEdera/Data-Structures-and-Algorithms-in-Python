
# 01 Creating a Tuple
### Time Complexity : $O(1)$
### Space Complexity : $O(N)$

```python
t = "a","b","c","d","e"
print(type(t))
print(t)
```

```python
t = ("a","b","c","d","e")
print(type(t))
print(t)
```

```python
t = ("a")
print(type(t))
print(t)
```

```python
t = ("a",)
print(type(t))
print(t)
```

```python
t = (1.0)
print(type(t))
print(t)
```

```python
t = (1.0,)
print(type(t))
print(t)
```

```python
t = tuple()
print(type(t))
print(t)
```

```python
t = tuple("abcdef")
print(type(t))
print(t)
```

# 04 Traverse a tuple
### Time Complexity : $O(N)$
### Space Complexity : $O(1)$

```python
t = ("a","b","c","d","e")
for i in t:
    print(i)
```

```python
t = ("a","b","c","d","e")
for i in range(len(t)):
    print(i,t[i])
```

> # Linear serach

### Time Complexity : $O(N)$
### Space Complexity : $O(1)$

```python
def searchInTuple(pTuple, element):
    for i in pTuple:
        if i == element:
            return pTuple.index(i)
    return 'The element does not exist'

```

```python
t = ("a","b","c","d","e")
print(searchInTuple(t, 'c'))
```
