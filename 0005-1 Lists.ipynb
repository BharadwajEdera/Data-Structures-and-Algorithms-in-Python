# 02 Inserting into List
> ### insert() : Time Complexity :: $O(N)$

> ### append() : Time Complexity :: $O(1)$ , Space Complexity :: $O(1)$ 

> ### extend() : Time Complexity :: $O(N)$ , Time Complexity :: $O(N)$ 



```python
# Insert - List 


myList.insert(4,15)
print(myList)

myList.append(55)
print(myList)

myList.append([60,65,70])
print(myList)

newList = [11,12,13,14]
myList.extend(newList)
print(myList)
```

# 04 Delete an Element in a List


### pop() without index deletes Last Element
> Time Complexity :: $O(1)$ , Space Complexity :: $O(1)$ 

### pop(index) with index returns the deleted item and removes from the list
> Time Complexity :: $O(N)$ , Space Complexity :: $O(1)$ 


```python
### pop() without index deltes Last Element

myList = [60,65,70,75,80,85,90,95]
myList.pop()
print(myList)
```

```python
### pop() with index returns the deleted item and removes from the list

myList = [60,65,70,75,80,85,90,95]
print(myList.pop(4))
print(myList)
```

### delete() 
### to delete Multiple Elements in List

> Time Complexity :: $O(N)$ , Space Complexity :: $O(1)$ 

```python
### delete()

myList = [60,65,70,75,80,85,90,95]
del myList[4]
print(myList)

### delete() multiple elements

myList = [60,65,70,75,80,85,90,95]
del myList[4:]
print(myList)
```

# 05  Searching for an element in the List


### IN operator : It checks each and every elemnt
> Time Complexity :: $O(N)$ , Space Complexity :: $O(1)$ 

```python
#  Searching for an element in the List
myList =  [10,20,30,40,50,60,70,80,90]

if 20 in myList:
    print(myList.index(20))

else:
    print("The Value Doesnt Exists")
```

### Linear Search : It checks each and every elemnt
> Time Complexity :: $O(N)$ , Space Complexity :: $O(1)$ 

```python
#  Searching for an element in the List
myList =  [10,20,30,40,50,60,70,80,90]

def searchinList(list, value):
    for i in list:
        if i == value:
            return list.index(value)
    return 'The value does not exist in the list'

print(searchinList(myList, 100))
```

```python
#  List operations / functions
total = 0 
count = 0
while (True):
    inp = input('Enter a number: ') 
    if inp == 'done': break
    value = float(inp)
    total = total + value
    count = count + 1 
    average = total / count

print('Average:', average)
```

```python
numlist = list() 
while (True):
    inp = input('Enter a number: ') 
    if inp == 'done': break
    value = float(inp)
    numlist.append(value)

average = sum(numlist) / len(numlist) 
print('Average:', average)
```
