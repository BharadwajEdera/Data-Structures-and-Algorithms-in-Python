# L_O_T : queue with List

```python
class Queue:
    def __init__(self):
        self.List = []
    def __str__(self):
        values = [str(x) for x in self.List]
        return "\n".append(values)
    def __iter__(self):
        for i in self.List:
            yield i
            
    def isEmpty(self):
        if self.List == []:
            return True
        else:
            return False
    
            
    def enqueue(self,value):
        self.List.append(value)
        
    def dequeue(self):
        return self.List.pop(0)
        
        
```

```python
def L_O_T(rootNode):
    if not rootNode:
        return 
    q = Queue()
    q.enqueue(rootNode)
    while not(q.isEmpty()):
        root = q.dequeue()
        print(root.data)
        if (root.LeftChild is not None):
            q.enqueue(root.LeftChild)
        if (root.RightChild is not None):
            q.enqueue(root.RightChild)
```

```python
L_O_T(T)
```

# Search for a Node in Tree

```python
def SesrchNode(rootNode,Node):
    if not rootNode:
        return 
    q = Queue()
    q.enqueue(rootNode)
    while not(q.isEmpty()):
        root = q.dequeue()
        if root.data == Node:
            return "Element Found"
        if (root.LeftChild is not None):
            q.enqueue(root.LeftChild)
        if (root.RightChild is not None):
            q.enqueue(root.RightChild)
    return "Element Not Found"
    
```

```python
SesrchNode(T,"Cold")
```

```python
SesrchNode(T,"Choclate")
```

# Insert_Into_tree

```python
def Insert_into_Tree(rootNode,newNode):
    if not rootNode:
        rootNode = newNode
    q = Queue()
    q.enqueue(rootNode)
    while not(q.isEmpty()):
        root = q.dequeue()

        if (root.LeftChild is not None):
            q.enqueue(root.LeftChild)
        else:
            root.LeftChild = newNode
            return "Successfully Inserted"
            
            
        if (root.RightChild is not None):
            q.enqueue(root.RightChild)
    
        else:
            root.RightChild = newNode
            return "Successfully Inserted"

    
```

>> ## Get Deepest Node

```python
def Get_deepest(rootNode):
    if not rootNode:
        return 
    q = Queue()
    q.enqueue(rootNode)
    while not(q.isEmpty()):
        root = q.dequeue()

        if (root.LeftChild is not None):
            q.enqueue(root.LeftChild)
        if (root.RightChild is not None):
            q.enqueue(root.RightChild)
            
    deepestNode = root
    return deepestNode
```

>> # Delete DeepestNode

```python
def Delete_deepest(rootNode,deepestNode):
    if not rootNode:
        return 
    q = Queue()
    q.enqueue(rootNode)
    while not(q.isEmpty()):
        root = q.dequeue()


        if (root.LeftChild is not None):
            if root.LeftChild == deepestNode:
                root.LeftChild = None
                return "Deleted"
            
            else:
                q.enqueue(root.LeftChild)


        if (root.RightChild is not None):
            if root.RightChild == deepestNode:
                root.RightChild = None
                return "Deleted"
            
            else:
                q.enqueue(root.LeftChild)
```
