```python
def power(base, exponent):
    
    assert int(exponent) == exponent , "exponent should be Integer"
    
    if exponent == 0:
        return 1
    if exponent == 1:
        return base
    
    else:
        return base * power( base , exponent-1 )    
```

```python
power(2, 0)
```

```python
power(2, 2)
```

```python
power(2, 4)
```

```python
def factorial(num):
    assert num>=0 and int(num) == num , "num must be Positive Integer"
    
    if num in (0,1):
        return 1
    else:
        return num * factorial(num-1)        
```

```python
factorial(1)
```

```python
factorial(2)
```

```python
factorial(4)
```

```python
def productOfArray(arr):
    
    if len(arr) == 0:
        return 1
    
    else:
        return arr[len(arr)-1] * productOfArray(arr[:len(arr)-1])
```

```python
productOfArray([1,2,3])
```

```python
productOfArray([1,2,3,10])
```

```python
def recursiveRange(num):
    
    if num == 0:
        return 0
    
    else:
        return num + recursiveRange(num-1)
```

```python
recursiveRange(6)
```

```python
recursiveRange(10)
```

```python
def fib(num):
    #assert num >=0 and int(num) == num , "Num should be positive integer"
    if num in (0,1):
        return num
    else:
        return fib(num-1) + fib(num-2)
```

```python
fib(4)
```

```python
fib(10)
```

```python
def reverse(strng):
    
    if len(strng) == 1:
        return strng[0]
    else:
        return strng[len(strng)-1] + reverse(strng[:(len(strng)-1)])
```

```python
def someRecursive(arr, cb):
    if len(arr) == 0:
        return False
    if not(cb(arr[0])):
        return someRecursive(arr[1:], cb)
    else:
        return True

def isOdd(num):
    if num%2==0:
        return False
    else:
        return True
```

```python
someRecursive([1,2,3,4], isOdd)
```

```python
someRecursive([4,6,8,9], isOdd)
```

```python
someRecursive([4,6,8], isOdd)
```

```python
def flatten(arr):
    if arr == []:
        return arr
    else:
        if type(arr[0]) == list :
            return flatten(arr[0]) + flatten(arr[1:])
        else :
            return arr[:1] + flatten(arr[1:])
```

```python
flatten([1,2,3,[4,5]])
```

```python
flatten([1,[2,[3,4],[[5]]]])
```

```python
flatten([[1],[2],[3]])
```

```python
def nestedEvenSum(obj, sum=0):
    for i in obj.keys():
        if type(obj[i]) == int and (obj[i] % 2) == 0:
            sum += obj[i]
        elif type(obj[i]) == dict:
            sum += nestedEvenSum(obj[i])
    return sum
```

```python
obj1 = { "outer": 2, "obj": { "inner": 2, "otherObj": { "superInner": 2, "notANumber": True, "alsoNotANumber": "yup" } } }
nestedEvenSum(obj1)
```

```python
def capitalizeWords(arr):
    if arr == []:
        return arr
    else:
        return [arr[0].upper()] + capitalizeWords(arr[1:])
```

```python
arr = ['i', 'am', 'learning', 'recursion']
capitalizeWords(arr)
```

```python
def stringifyNumbers(obj):
    
    for i in obj.keys():
        if type(obj[i]) == int:
            obj[i] = str(obj[i])
        elif type(obj[i]) == dict:
            obj[i] = stringifyNumbers(obj[i])
            
    return obj
```

```python
obj = {"num":1, "test":[], "data": {"val":4, "info":{"isRight": True, "random": 66}}}
stringifyNumbers(obj)
```

```python
def collectStrings(obj, l = []):
    #print(obj)
    for i in obj.keys():
        if type(obj[i]) == str:
            l.append(obj[i])
            #print(l)
        elif type(obj[i]) == dict:
            #print(obj[i])
            obj[i] = collectStrings(obj[i],l)
        else:
            continue
    return l
```

```python
obj1 = { "outer": 2, "obj": { "inner": 2, "otherObj": { "superInner": 2, "notANumber": True, "alsoNotANumber": "yup" } } }

collectStrings(obj1)
```
