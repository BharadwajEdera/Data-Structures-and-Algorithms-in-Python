# Question 1 : Single Python List to Implement Three Stacks


```python
class Multistack:
    def __init__(self,stackSize):
        self.number_stacks = 3
        self.List = [0] * stackSize * self.number_stacks
        self.Starting_Sizes = [0] + [i*stackSize for i in range(1,self.number_stacks+1)]
        self.Sizes = [0] + [i*stackSize for i in range(1,self.number_stacks)]
        self.stackSize = stackSize
        
    def __str__(self):
        values = [str(i) for i in self.List]
        Starting_Sizes = [str(i) for i in self.Starting_Sizes]
        Sizes = [str(i) for i in self.Sizes]
        return "List :" + " ".join(values) + "\n" + "Starting_Sizes :" + " ".join(Starting_Sizes) + "\n" + "Sizes :" + " ".join(Sizes)

       
    def isEmpty(self,stack_number):
        if self.Sizes [stack_number] == self.Starting_Sizes[stack_number]:
            return True
        else:
            return False
        
    def isFull(self,stack_number):
        if  self.Sizes[stack_number] == self.Starting_Sizes[stack_number+1]:
            return True
        else:
            return False
    
    
    def push(self,value,stack_number):
        if self.isFull(stack_number):
            return f"{stack_number} : is Full"
        else:
            self.List[self.Sizes[stack_number]] = value
            self.Sizes[stack_number] += 1
            return f"{value} Element inserted in {stack_number} stack"
        
    def pop(self,stack_number):
        if self.isEmpty(stack_number):
            return "Stack is Empty"
        else:
            element = self.List[self.Sizes[stack_number]-1]
            self.List[self.Sizes[stack_number]-1] = 0
            self.Sizes[stack_number] -= 1
            return f"{element} got deleted in {stack_number} stack"
        
    def peek(self,stack_number):
        if self.isEmpty(stack_number):
            return "Stack is Empty"  
        else:
            return self.List[self.Sizes[stack_number]-1]
      
```

# Implement Stack with push , pop and min methods each operates at $O(1)$


> ## Hint :: Create Two Linked Lists : 1st for Stack , 2nd for min() 

> ## Pass minNode.next and top.next through argument to Node class

```python
class Node():
    def __init__(self, value=None, next = None):
        self.value = value
        self.next = next
    
    def __str__(self):
        string = str(self.value)
        if self.next:
            string += ',' + str(self.next)
        return string

class Stack():
    def __init__(self):
        self.top = None
        self.minNode = None
    
    def min(self):
        if not self.minNode:
            return None
        return self.minNode.value
    
    def push(self, item):
        if self.minNode and (self.minNode.value < item):
            self.minNode = Node(value = self.minNode.value, next=self.minNode)
        else:
            self.minNode = Node(value = item, next=self.minNode)
        self.top = Node(value=item, next=self.top)
    
    def pop(self):
        if not self.top:
            return None
        self.minNode = self.minNode.next
        item = self.top.value
        self.top = self.top.next
        return item

```

```python
customStack = Stack()
customStack.push(5)
print(customStack.min())
customStack.push(6)
print(customStack.min())
customStack.push(3)
print(customStack.min())
customStack.pop()
print(customStack.min())
```

# Question 3: Stack of plates


```python
class Stack_of_plates:
    def __init__(self,maxPlates):
        self.maxPlates = maxPlates
        self.list = []
        
    def __str__(self):
        if len(self.list) == 0:
            return "Empty Stacks"
        else:
            values = [str(x) for x in  self.list ]
            return " ".join(values)
    
    def push(self,value):
        if len(self.list) > 0 and (len(self.list[-1]) < self.maxPlates):
            self.list[-1].append(value)
        else:
            self.list.append([value])
            
    def pop(self):
        while len(self.list) > 0 and len(self.list[-1]) == 0:
            self.list.pop()
        if len(self.list) == 0:
            return "Stack is Empty"
        else:
            return self.list[-1].pop()
        
    def pop_at(self,stack_number):
        if len(self.list) == [] :
            return "Empty Stack"
        else:
            return self.list[stack_number].pop()
            
            
            
    
```

```python
class Stack():
    def __init__(self):
        self.list = []
        
    def __iter__(self):
        for i in self.list:
            yield i
  
    def __len__(self):
        return len(self.list)
  
    def push(self, item):
        self.list.append(item)
  
    def pop(self):
        if len(self.list) == 0:
            return None
        return self.list.pop()

class QueueviaStack():
    def __init__(self):
        self.inStack = Stack()
        self.outStack = Stack()
        
    def __str__(self):
        values = [str(x) for x in self.inStack]
        return " ".join(values)
  
    def enqueue(self, item):
        self.inStack.push(item)
  
    def dequeue(self):
        while len(self.inStack):
            self.outStack.push(self.inStack.pop())
        result = self.outStack.pop()
        while len(self.outStack):
            self.inStack.push(self.outStack.pop())
        return result
```

```python
q = QueueviaStack()
```

```python
q.enqueue(10)
q.enqueue(20)
q.enqueue(30)
q.enqueue(40)
print(q)
```

```python
class AnimalShelter():
    def __init__(self):
        self.cats = []
        self.dogs = []
  
    def enqueue(self, animal, type):
        if type == 'Cat':
            self.cats.append(animal)
        else:
            self.dogs.append(animal)
    
    def dequeueCat(self):
        if len(self.cats) == 0:
            return None
        else:
            cat = self.cats.pop(0)
        return cat

    def dequeueDog(self):
        if len(self.dogs) == 0:
            return None
        else:
            dog = self.dogs.pop(0)
        return dog

    def dequeueAny(self):
        if len(self.cats) == 0:
            result = self.dogs.pop(0)
        else:
            result = self.cats.pop(0)
        return result
```

```python
customQueue = AnimalShelter()
customQueue.enqueue('Cat1', 'Cat')
customQueue.enqueue('Cat2', 'Cat')
customQueue.enqueue('Dog1', 'Dog')
customQueue.enqueue('Cat3', 'Cat')
customQueue.enqueue('Dog2', 'Dog')
print(customQueue.dequeueAny())
```
