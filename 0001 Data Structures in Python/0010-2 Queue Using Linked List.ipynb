# Using Linked List
> ## Enqueue : at the End
> ## Dequeue : at the Beginning

```python
class Node:
    def __init__(self, value=None):
        self.value = value
        self.next = None
    
    def __str__(self):
        return str(self.value)

class LinkedList:
    def __init__(self):
        self.head = None
        self.tail = None
    
    def __iter__(self):
        curNode = self.head
        while curNode:
            yield curNode
            curNode = curNode.next

class Queue:
    def __init__(self):
        self.linkedList = LinkedList()
    
    def __str__(self):
        if self.linkedList.head == None:
            return "Queue is EMpty"
        else:
            values = [str(x) for x in self.linkedList]
            return ' '.join(values)
    
    def enqueue(self, value):
        newNode = Node(value)
        if self.linkedList.head == None:
            self.linkedList.head = newNode
            self.linkedList.tail = newNode
        else:
            self.linkedList.tail.next = newNode
            self.linkedList.tail = newNode
    
    def isEmpty(self):
        if self.linkedList.head == None:
            return True
        else:
            return False
    
    def dequeue(self):
        if self.isEmpty():
            return "There is not any node in the Queue"
        else:
            tempNode = self.linkedList.head
            if self.linkedList.head == self.linkedList.tail:
                self.linkedList.head = None
                self.linkedList.tail = None
            else:
                self.linkedList.head = self.linkedList.head.next
            return tempNode
    
    def peek(self):
        if self.isEmpty():
            return "There is not any node in the Queue"
        else:
            return self.linkedList.head
    
    def delete(self):
        self.linkedList.head = None
        self.linkedList.tail = None
        return "Queue Deleted"
```

```python
q = Queue()
print(q)
```

```python
q.enqueue(1)
q.enqueue(2)
q.enqueue(3)
q.enqueue(4)
```

```python
print(q)
```

```python
print(q.dequeue())
```

```python
print(q)
```

```python
print(q.peek())
```

```python
q.delete()
```

# Implementation of Queue using Python Modules

# 01. Python collections.deque Class


```python

from collections import deque

customQueue = deque(maxlen=3)
print(customQueue)
```

```python

customQueue.append(1)
customQueue.append(2)
customQueue.append(3)
customQueue.append(4)
print(customQueue)
```

```python
print(customQueue.popleft())
```

```python
print(customQueue)
```

```python
print(customQueue.clear())
```

```python
print(customQueue)
```

# 02. Python queue Module


```python
import queue as q

customQueue = q.Queue(maxsize=3)
print(customQueue.empty())
```

```python
print(customQueue.qsize())
```

```python
customQueue.put(1)
customQueue.put(2)
customQueue.put(3)
```
