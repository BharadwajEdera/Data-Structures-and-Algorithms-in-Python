### Calculate the Average Temperature Without List

```python
numdays = int(input("Enter number of days"))
total = 0
for i in range(1,numdays+1):
    nextday = int(input("Day"+str(i)+"'s High temperature'"))
    total += nextday

avg = round(total/numdays,2)
print("\nAverage = "+str(avg))
```

### Calculate the Average Temperature With List

```python
numdays = int(input("Enter number of days"))
total = 0
l = []
for i in range(0,numdays):
    nextday = int(input("Day"+str(i+1)+"'s High temperature'"))
    l.append(nextday)
    total += l[i]
    
avg = round(total/numdays,2)
print("\nAverage = "+str(avg))

above=0
for i in l:
    if i > avg:
        above += 1
print(str(above)+"day(s) above average")
        
```

### Question 2



### Note (2,2) or (3,3) not Valid Pairs

```python
def pairSum(myList, sum):
    l = []
    for i in range(len(myList)):
        for j in range(i+1,len(myList)):
            if (myList[i]+myList[j]) == sum:
                l.append(str(myList[i])+ "+" +str(myList[j]))
            else:
                continue
    return l
```

```python
pairSum([2,4,3,5,6,-2,4,7,8,9], 7)
```

### Question 4  Find Maximum Product of Two integers in an array where all elemnts are positive

```python
def findMaxProduct(l):
    max = 0
    for i in range(len(l)):
        for j in range(i+1,len(l)):
            if l[i]*l[j] > max:
                max = l[i]*l[j] 
                pairs = str(l[i]) + "*" + str(l[j])
    
    print(f"{pairs} = {max}")
                    
    return max
```

```python
findMaxProduct([2,4,3,5,6,2,4,7,8,9])
```

### Question To find If all elements in kist or array are unique


```python
def isUnique(l):
    r = []
    for i in range(len(l)):
        if l[i] not in r:
            r.append(l[i])
        else:
            return "Given List has Duplicate Elements"
    return "Given List has Unique Elements"
```

```python
isUnique([2,4,3,5,6,2,4,7,8,9])
```

### Question 7


```python
def rotate_matrix(matrix):
    '''rotates a matrix 90 degrees clockwise'''
    n = len(matrix)
    for layer in range(n // 2):
        first, last = layer, n - layer - 1
        for i in range(first, last):
            # save top
            top = matrix[layer][i]

            # left -> top
            matrix[layer][i] = matrix[-i - 1][layer]

            # bottom -> left
            matrix[-i - 1][layer] = matrix[-layer - 1][-i - 1]

            # right -> bottom
            matrix[-layer - 1][-i - 1] = matrix[i][- layer - 1]

            # top -> right
            matrix[i][- layer - 1] = top
    return matrix

matrix = [[1,2], [3,4]]
print(matrix)
print(rotate_matrix(matrix))
```
