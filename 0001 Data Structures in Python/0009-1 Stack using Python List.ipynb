# Creating Stack without size Limit :: By using List

## For creating a Empty stack
### Time Complexity :: $O(1)$
### Space Complexity :: $O(1)$
## For creating a isEmpty() method
### Time Complexity :: $O(1)$
### Space Complexity :: $O(1)$

## For creating a push() method using append() builtin function
### Time Complexity :: amortized  $O(n)$
### Space Complexity :: $O(1)$
## For creating a pop() method using pop() builtin function
### Time Complexity :: $O(1)$
### Space Complexity :: $O(1)$
## For creating a peek() method 
### Time Complexity :: $O(1)$
### Space Complexity :: $O(1)$


```python
class stack:
    def __init__(self):
        self.list = []
        
    def __str__(self):
        values = reversed(self.list)
        values = [str(x) for x in values]
        return "\n".join(values)
    
    # isEmpty
    def isEmpty(self):
        if self.list == []:
            return True
        elif self.list != []:
            return False
            
    def push(self,value):
        self.list.append(value)
        return "The element is successfully inserted"
    
    def pop(self):
        if self.isEmpty():
            return "there is no element in Stack"
        else:
            return self.list.pop()
        
    def peek(self):
        if self.isEmpty():
            return "there is no element in Stack"
        else:
            return self.list[-1]
        
    def delete(self):
        self.list = None
        return "Stack deleted"
```

```python
s.push(1)
s.push(2)
s.push(3)
print(s)
```

```python
print(s.isEmpty())
```

```python
print(s.pop())
```

```python
print(s)
```

```python
print(s)
```

```python
s.delete()
```

# Creating Stack with size Limit


## For creating a Empty stack
### Time Complexity :: $O(1)$
### Space Complexity :: $O(1)$
## For creating a isEmpty() method
### Time Complexity :: $O(1)$
### Space Complexity :: $O(1)$
## For creating a isFull() method
### Time Complexity :: $O(1)$
### Space Complexity :: $O(1)$
## For creating a push() method using append() builtin function
### Time Complexity :: amortized  $O(n)$
### Space Complexity :: $O(1)$
## For creating a pop() method using pop() builtin function
### Time Complexity :: $O(1)$
### Space Complexity :: $O(1)$
## For creating a peek() method 
### Time Complexity :: $O(1)$
### Space Complexity :: $O(1)$


```python
class stack:
    def __init__(self,maxSize):
        self.maxSize = maxSize
        self.list = []
        
    def __str__(self):
        values = reversed(self.list)
        values = [str(x) for x in values]
        return "\n".join(values)
    
    # isEmpty
    def isEmpty(self):
        if self.list == []:
            return True
        elif self.list != []:
            return False
        
    def isFull(self):
        if len(self.list) == self.maxSize:
            return True
        else:
            return False
    
    def push(self,value):
        if self.isFull():
            return "Stack is Full"
        else:
            self.list.append(value)
            return "Element is successfully inserted into stack"
        
    def pop(self):
        if self.isEmpty():
            return "there is no element in Stack"
        else:
            return self.list.pop()
        
    def peek(self):
        if self.isEmpty():
            return "there is no element in Stack"
        else:
            return self.list[-1]
        
    def delete(self):
        self.list = None
        return "Stack deleted"        
```

```python
s = stack(5)
s.isEmpty()
```

```python
s.push(1)
s.push(2)
s.push(3)
s.push(4)
```
