# 01 Creating a Circular Single Linked List with only one Node
### Time Complexity : $O(1)$
### Space Complexity : $O(1)$


```python
class Node:
    def __init__(self, value=None):
        self.value = value
        self.next = None  ### Address of Next Node
        
class CSLL:
    def __init__(self):
        self.head = None
        self.tail = None
        
    def display(self):
        
        if self.head is None:
            print("Empty Linked List")
        
        else:
            temp = self.head
            index = 1
            while temp:
                print(temp.value,"----->",end=" ")
                if self.head == self.head.next:
                    print(" ")
                    print("^      |")
                    print("|      |")
                    print("<-------")
                    break
                elif temp.next == self.head:
                    print(" ")
                    style = "<-------"
                    style1 = "        "
                    print("|"+style1*index+"|")
                    print(style*index)
                    break
                        
                temp = temp.next
                index += 1 
```

```python
L = CSLL()
L.display()
```

```python
node1 = Node(10)
L.head = node1
L.head.next = node1
L.display()
```

# 02 Insertion Into CSLL


```python
class Node:
    def __init__(self, value=None):
        self.value = value
        self.next = None  ### Address of Next Node
        
class CSLL:
    def __init__(self):
        self.head = None
        self.tail = None
        
    def insert_beginning(self,value):
        newNode = Node(value)
        newNode.next = self.head
        temp = self.head
        while temp.next != self.head:
            temp = temp.next
        temp.next = newNode
        self.head = newNode
        
    def insert_position(self,value,position):
        newNode = Node(value)
        temp = self.head
        
        for i in range(position-1):
            temp = temp.next
            
        newNode.next = temp.next
        temp.next = newNode
        
    def insert_end(self,value):
        newNode = Node(value)
        newNode.next = self.head
        temp = self.head
        while self.head != temp.next:
            temp = temp.next
            
        temp.next = newNode
        
    
        
    def display(self):
        
        if self.head is None:
            print("Empty Linked List")
        
        else:
            
            temp = self.head
            index = 1
            
            while temp:
                print(temp.value,"----->",end=" ")
                if self.head == self.head.next:
                    print(" ")
                    print("^      |")
                    print("|      |")
                    print("<-------")
                    break
                elif temp.next == self.head:
                    print(" ")
                    style = "<-------" 
                    style1 = "         "
                    print("|"+style1*(index)+"|")
                    print(style*(index+1))
                    break
                        
                temp = temp.next
                index += 1 
```

```python
L = CSLL()

node1 = Node(10)
L.head = node1

node2 = Node(20)
node1.next = node2

node3 = Node(30)
node2.next = node3

node4 = Node(40)
node3.next = node4

node5 = Node(50)
node4.next = node5

node5.next = L.head

L.display()
```

# 03 Search an element in CSLL

```python
class Node:
    def __init__(self, value=None):
        self.value = value
        self.next = None  ### Address of Next Node
        
class CSLL:
    def __init__(self):
        self.head = None
        self.tail = None
        
        
    def searchCSLL(self,value):
        temp = self.head
        while temp.next != self.head:
            if temp.value == value:
                return value
            temp = temp.next
        return "Element Doesn't exist in CSLL"
        
    def display(self):
        
        if self.head is None:
            print("Empty Linked List")
        
        else:
            temp = self.head
            index = 1
            while temp:
                print(temp.value,"----->",end=" ")
                if self.head == self.head.next:
                    print(" ")
                    print("^      |")
                    print("|      |")
                    print("<-------")
                    break
                elif temp.next == self.head:
                    print(" ")
                    style = "<-------"
                    style1 = "        "
                    print("|"+style1*index+"|")
                    print(style*index)
                    break
                        
                temp = temp.next
                index += 1 
```

```python
class Node:
    def __init__(self, value=None):
        self.value = value
        self.next = None  ### Address of Next Node
        
class CSLL:
    def __init__(self):
        self.head = None
        self.tail = None
        
    def delete_beginning(self):
        temp = self.head
        while temp.next != self.head:
            temp = temp.next
        temp.next = self.head.next
        self.head = self.head.next
        
    def delete_postion(self,position):
        prev = self.head
        temp = self.head.next
        for i in range(position-1):
            temp = temp.next
            prev = prev.next
            
        prev.next = temp.next
        
    def delete_end(self):
        prev = self.head
        temp = self.head.next
        while temp.next != self.head:
            temp = temp.next
            prev = prev.next
            
        prev.next = self.head
        

        
    def display(self):
        
        if self.head is None:
            print("Empty Linked List")
        
        else:
            temp = self.head
            index = 1
            while temp:
                print(temp.value,"----->",end=" ")
                if self.head == self.head.next:
                    print(" ")
                    print("^      |")
                    print("|      |")
                    print("<-------")
                    break
                elif temp.next == self.head:
                    print(" ")
                    style = "<-------"
                    style1 = "        "
                    print("|"+style1*index+"|")
                    print(style*index)
                    break
                        
                temp = temp.next
                index += 1 
```
