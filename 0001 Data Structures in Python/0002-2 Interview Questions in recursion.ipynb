### Question 1: Sum of Ddigits of Positive integers using Recursion


```python
def sum_of_dig(n):
    assert n>=0 and int(n) == n , "The number Should be Positive interger"
    if n in (0,1):
        return n
    else:
        return int(n%10) + sum_of_dig(int(n/10))
```

```python
sum_of_dig(112)
```

### Question 2: How to Calculate power of a number using Recursion


```python
def power(x,n):
    assert n>=0 and int(n) == n , "The exponent should be positive integer only"
    if n== 0:
        return 1
    elif n == 1:
        return x
    else:
        return x * power(x , (n-1))
```

```python
power(2,3)
```

### Question 3 : How to find Greates Common Divisor of Two numbers using Recursion


```python
def GCD(a,b):
    assert int(a) == a and int(b) == b , "Number must be Integer"
    if a<0 or b<0:
        a = abs(a) 
        b = abs(b)
        
    if b == 0:
        return a
    else:
        return GCD(b , a%b)
    
```

```python
GCD(8,12)
```

```python
GCD(48,18)
```

### Question 4 : How to Convert a number from decimal to binary using recursion


```python
def Decimal_to_Binary(n):
    assert int(n)==n , "The number should be interger"
    if n==0 :
        return 0
    
    else:
        return n%2 + 10 * Decimal_to_Binary(int(n/2))
```

```python
Decimal_to_Binary(10)
```

```python
Decimal_to_Binary(14)
```

```python
Decimal_to_Binary(-14)
```

```python
Decimal_to_Binary(3.5)
```
