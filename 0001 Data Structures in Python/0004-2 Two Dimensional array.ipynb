# 01 Creating two dimensional Array

### Creating Two Dimensional array : Time Complexity :: $O(N)$
### Creating Empty Two Dimensional array and Inserting values : Time Complexity :: $O(M*N)$
>M : Number of rows

> N : Number of Columns

```python
import numpy as np
```

```python
twoDArray = np.array([[11, 15, 10, 6], [10, 14, 11, 5], [12, 17, 12, 8], [15, 18, 14, 9] ])
print(twoDArray)
```

# 02 Inserting Elements into wo dimensonal array




###  Time Complexity :: $O(M*N)$
>M : Number of rows

> N : Number of Columns

### Inserting At the end of Array Time Complexity :: $O(1)$




```python
twoDArray = np.array([[11, 15, 10, 6], [10, 14, 11, 5], [12, 17, 12, 8], [15, 18, 14, 9] ])
print(twoDArray)
print(" ")
newTwoDArray = np.insert(twoDArray, 1, [[1,2,3,4]], axis=1)
print("2d array after adding a Column")
print(newTwoDArray)
```

```python
twoDArray = np.array([[11, 15, 10, 6], [10, 14, 11, 5], [12, 17, 12, 8], [15, 18, 14, 9] ])
print(twoDArray)
print(" ")
newTwoDArray = np.insert(twoDArray, 1, [[1,2,3,4]], axis=0)
print("2d array after adding a Row")
print(newTwoDArray)
```

```python
### Adding a Row/Column at end 
twoDArray = np.array([[11, 15, 10, 6], [10, 14, 11, 5], [12, 17, 12, 8], [15, 18, 14, 9] ])
print(twoDArray)
print(" ")
newTwoDArray = np.append(twoDArray, [[1,2,3,4]], axis=0)
print("2d array after adding a Row")
print(newTwoDArray)
```



###  Time Complexity :: $O(1)$


### Space Complexity :: $O(1)$

```python
def accessElements(array, rowIndex, colIndex):
    if rowIndex >= len(array) and colIndex >= len(array[0]):
        print('Incorrect Index')
    else:
        print(array[rowIndex][colIndex])

accessElements(newTwoDArray, 1, 2)
```



###  Time Complexity :: $O(M*N)$


### Space Complexity :: $O(1)$

```python
def traverseTDArray(array):
    for i in range(len(array)):
        for j in range(len(array[0])):
            print(array[i][j])
            
traverseTDArray(twoDArray)
```



###  Time Complexity :: $O(M*N)$


### Space Complexity :: $O(1)$

```python
def searchTDArray(array, value):
    for i in range(len(array)):
        for j in range(len(array[0])):
            if array[i][j] == value:
                return 'The value is located index '+str(i)+" "+str(j)
    return 'The element no found'


print(searchTDArray(twoDArray, 444))
```

```python
twoDArray = np.array([[11, 15, 10, 6], [10, 14, 11, 5], [12, 17, 12, 8], [15, 18, 14, 9] ])
print(twoDArray)
print(" ")


newTDArray = np.delete(twoDArray, 1, axis=1)
print(newTDArray)
```
