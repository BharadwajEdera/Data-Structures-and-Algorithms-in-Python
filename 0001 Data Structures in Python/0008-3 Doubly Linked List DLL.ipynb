# Creation of DLL


```python
class Node:
    def __init__(self,value):
        self.value = value
        self.next = None
        self.before = None
        
class DLL:
    def __init__(self):
        self.head = None
        
    def display(self):
        
        if self.head == None:
            return "DLL is Empty"
        
        else:
            temp = self.head
            while temp:
                print(temp.value,"<====>",end = " ")
                temp = temp.next
                
```

```python
class Node:
    def __init__(self,value):
        self.value = value
        self.next = None
        self.before = None
        
class DLL:
    def __init__(self):
        self.head = None
        
        
    def insert_beginning(self,value):
        newNode = Node(value)
        newNode.next = self.head
        self.head.before = newNode
        self.head = newNode
        
    def insert_end(self,value):
        newNode = Node(value)
        temp = self.head
        while temp.next:
            temp = temp.next
        temp.next = newNode
        newNode.before = temp
        
    def insert_position(self,value,position):
        newNode = Node(value)
        temp = self.head
        for i in range(position-1):
            temp = temp.next
            
        newNode.before = temp
        newNode.next = temp.next
        temp.next.before = newNode
        temp.next = newNode
        
    def insert_end(self,value):
        newNode = Node(value)
        temp = self.head
        while temp.next:
            temp = temp.next
        
        newNode.before = temp
        temp.next = newNode
        
    
    def search_DLL(self,value):
        temp = self.head
        while temp:
            if temp.value == value:
                return value
            temp = temp.next
        return "Element Doesn't exists in DLL"
    
    
    def delete_beginning(self):
        temp = self.head
        temp.next.before = None
        self.head = temp.next
        
    def delete_position(self,position):
        prev = self.head
        temp = self.head.next
        
        for i in range(position-1):
            temp = temp.next
            prev = prev.next
            
        prev.next = temp.next
        temp.next.before = prev
        
    def delete_end(self):
        prev = self.head
        temp = self.head.next
        
        while temp.next:
            temp = temp.next
            prev = prev.next
            
        prev.next = None
        
    def delete_Entire_DLL(self):
        temp = self.head
        while temp.next:
            temp.before = None
            temp = temp.next
        self.head = None
            
    
    def Reverse_display(self):
        if self.head == None:
            return "DLL is Empty"
        
        else:
            temp = self.head
            while temp.next:
                temp = temp.next
            
            while temp:
                print(temp.value,"<=====>",end=" ")
                temp = temp.before
        
           
        
    def display(self):
        
        if self.head == None:
            return "DLL is Empty"
        
        else:
            temp = self.head
            while temp:
                print(temp.value,"<====>",end = " ")
                temp = temp.next
                
```
