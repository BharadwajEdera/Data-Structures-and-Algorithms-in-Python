# Search for a String 


```python
class TrieNode:
    def __init__(self):
        self.children = {}  # Dictionary
        self.endOfString = False # By default False
```

```python
class Trie:
    def __init__(self):
        self.root = TrieNode()
        
    def insertString(self, word):
        current = self.root
        for i in word:
            ch = i
            print(ch)
            node = current.children.get(ch)
            print(node)
            if node == None:
                node = TrieNode()
                current.children.update({ch:node})
                print(current.children)
            current = node
            print(current)
        current.endOfString = True
        print(self.root.children)
        print("Successfully inserted")
    
    def searchString(self, word):
        currentNode = self.root
        for i in word:
            node = currentNode.children.get(i)
            if node == None:
                return False
            currentNode = node

        if currentNode.endOfString == True:
            return True
        else:
            return False
```

```python
newTrie = Trie()
```

```python
newTrie.insertString("AIR")
newTrie.insertString("AIT")
newTrie.insertString("BAR")
newTrie.insertString("BIL")
newTrie.insertString("BM")
```

```python
print(newTrie.searchString("App"))
```

```python
print(newTrie.searchString("Ap"))
```

# Delete a String in Trie


```python
def deleteString(root, word, index):
    ch = word[index]
    currentNode = root.children.get(ch)
    canThisNodeBeDeleted = False

    if len(currentNode.children) > 1:
        deleteString(currentNode, word, index+1)
        return False
    
    if index == len(word) - 1:
        if len(currentNode.children) >= 1:
            currentNode.endOfString = False
            return False
        else:
            root.children.pop(ch)
            return True
    
    if currentNode.endOfString == True:
        deleteString(currentNode, word, index+1)
        return False

    canThisNodeBeDeleted = deleteString(currentNode, word, index+1)
    if canThisNodeBeDeleted == True:
        root.children.pop(ch)
        return True
    else:
        return False
```
