# 05 Traverse Through a Dictionary
> ### Time Complexity :: $O(N)$

> ### Space Complexity :: $O(1)$

```python
def traverseDict(dict):
    for key in dict:
        print(key, dict[key])
```

```python
my_dict = {"one":1,"two" : "2" , "Three" : 3 , "four" : 4}
traverseDict(my_dict)
```

# 06 Searching a dictionary
> ### Time Complexity :: $O(N)$

> ### Space Complexity :: $O(1)$

```python
def searchDict(dict, value):
    for key in dict:
        if dict[key] == value:
            return key, value
    return 'The value does not exist'
```

```python
my_dict = {"one":1,"two" : "2" , "Three" : 3 , "four" : 4}
print(searchDict(my_dict, 4))
```

# copy()
> # Returns a Shallow copy of a Dictionary



```python
my_dict = {"one":1,"two" : "2" , "Three" : 3 , "four" : 4}
print(my_dict)
dict_copy = my_dict.copy()
print(dict_copy)
```
