# delete() using Linked List


```python
class Node:
    def __init__(self,value):
        self.value = value
        self.next = None
    
class LinkedList:
    def __init__(self):
        self.head = None
        
    def __iter__(self):
        temp = self.head
        while temp:
            yield temp
            temp = temp.next
        

        
class stack:
    def __init__(self):
        self.L = LinkedList()
        
    def __str__(self):
        if self.L.head == None:
            return "Stack is Empty"
        else:
            values = [str(x.value) for x in self.L]
            return "\n".join(values)
        
        
    def isEmpty(self):
        if self.L.head is None:
            return True
        else:
            False
            
    def push(self,value):
        newNode = Node(value)
        if self.L.head == None:
            self.L.head = newNode
        else:
            newNode.next = self.L.head
            self.L.head = newNode
            
    def pop(self):
        if self.L.head == None:
            return "Stack is empty"
        else:
            nodeValue = self.L.head.value
            self.L.head = self.L.head.next
            return f"{nodeValue} is removed"
            
    def peek(self):
        if self.L.head == None:
            return "Stack is empty"
        else:
            return self.L.head.value    
        
        
    def delete(self):
        self.head = None
        return "The Stack is deleted"
            
            
            
    
    
```

```python
s = stack()
print(s)
```

```python
s.push(1)
s.push(2)
s.push(3)
s.push(4)
print(s)
```

```python
s.pop()
```

```python
print(s)
```

```python
s.peek()
```

```python
s.delete()
```
