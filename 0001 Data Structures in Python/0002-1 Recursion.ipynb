# Factorial using Recursion


### Step 1 : Recursive Case - the flow


```python
def factorial(n):
    return n * factorial(n-1)
```

```python
factorial(3)
```

### RecursionError: maximum recursion depth exceeded

> Function will call itself and , it will never stop

> The system will Crash , if we implement in an Application

> To prevent the Above Error , We go for Step 2

### Step 2 : Base Case - The Stopping Criterion


```python
def factorial(n):
    if n in (0,1):
        return 1
    else:
        return n * factorial(n-1)
```

```python
factorial(3)
```

> So We come over the Error we got in the first step , By using Base Case(Stopping Criterion) in the Seond Step

```python
factorial(-1)
```

```python
factorial(1.5)
```

> So if we give Negitive Numbers or Fractional Input to Recursion. Then It will throw an RecursionError: maximum recursion depth exceeded in comparison

> So to avoid this error , we use Step 3 . By Raising Custom Exceptions By using raise method

### Step 3 : Unintnentional case - the Constraint


```python
def factorial(n):
    assert n>=0 and int(n) == n ,"The Number Must be positive integer only"
    if n in (0,1):
        return 1
    else:
        return n * factorial(n-1)
```

```python
factorial(-1)
```

```python
factorial(1.5)
```

```python
factorial(4)
```

# Fibonacci Numbers - Recursion


```python
def fibonacci(n):
    assert n >=0 and int(n) == n , "The number should be Positive integer only"
    if n in (0,1):
        return n
    else:
        return fibonacci(n-1) + fibonacci(n-2)
```

```python
fibonacci(5)
```

```python
fibonacci(7)
```

```python
fibonacci(-1)
```

```python
fibonacci(-1.5)
```

```python
fibonacci(4)
```
