```python
class Node:
    def __init__(self,value):
        self.value = value
        self.next = None
        self.before = None
        
class CDLL:
    
    def __init__(self):
        self.head = None
        
    def insert_beginning(self,value):
        newNode = Node(value)
        temp = self.head
        temp.before = newNode
        newNode.next = self.head
        
        while temp.next != self.head:
            temp = temp.next
        newNode.before = temp
        temp.next = newNode
        self.head = newNode
        
    def insert_position(self,value,position):
        newNode = Node(value)
        temp = self.head
        for i in range(position-1):
            temp = temp.next
        newNode.next = temp.next
        temp.next.before = newNode
        newNode.before = temp
        temp.next = newNode
            
    
    def insert_end(self,value):
        newNode = Node(value)
        temp = self.head
        while temp.next != self.head:
            temp = temp.next
        newNode.before = temp
        newNode.next = self.head
        self.head.before = newNode
        temp.next = newNode
        
        
    def delete_beginning(self):
        temp = self.head
        while temp.next != self.head:
            temp = temp.next
        temp.next = self.head.next
        self.head.before = temp
        self.head = self.head.next
        
        
    def delete_position(self,position):
        prev = self.head
        temp = self.head.next
        for i in range(position-1):
            temp = temp.next
            prev = prev.next
        prev.next = temp.next 
        temp.next.before = prev
        
        
    def delete_end(self):
        temp = self.head
        while temp.next != self.head:
            temp = temp.next
        self.head.before = temp.before
        temp.before.next = self.head
        
        
    def reverse_display(self):
        temp = self.head
        while temp.next != self.head:
            temp = temp.next
        index = 1
        while temp != self.head:
            print(temp.value,"<====>",end=" ")
            if self.head == self.head.next:
                print(" ")
                print("^      |")
                print("|      |")
                print("<-------")
                break
            elif temp.before == self.head:
                print(" ")
                style = "<-------"
                style1 = "         "
                print("|"+style1*index+"|")
                print(style*(index+1))
                break
            temp = temp.before
            index += 1 

        
        
        
        
    def display(self):
        
        if self.head == None:
            return "Empty CDLL"
        else:
            temp = self.head
            index = 1
            while temp:
                print(temp.value,"<====>",end=" ")
                if self.head == self.head.next:
                    print(" ")
                    print("^      |")
                    print("|      |")
                    print("<-------")
                    break
                elif temp.next == self.head:
                    print(" ")
                    style = "<-------"
                    style1 = "         "
                    print("|"+style1*index+"|")
                    print(style*(index+1))
                    break
                        
                temp = temp.next
                index += 1 
```

```python
L = CDLL()
L.display()
```
